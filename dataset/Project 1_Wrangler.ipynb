```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
#!type resale_Lincoln_MKS.csv
```

```python
# load dataset
df_Wrangler = pd.read_csv('resale_Jeep_Wrangler.csv')
df_Wrangler.head()
```

```python
## clean the dataset
# drop rows that have NaN
df_Wrangler = df_Wrangler.dropna(axis=0, how='any')
```

```python
## Explore the dataset: Take care of the config: only care about Base and EcoBoost
#df_MKS.replace(' BASE', ' Base', inplace = True)
#df_MKS.replace(' Base', 'Base', inplace = True)
#df_MKS.replace(' EcoBoost', 'EcoBoost', inplace = True)
df_Wrangler = df_Wrangler[df_Wrangler['config'].isin(['Sport', 'Sahara', 'Rubican', 'X', 'Unlimited Sahara', 'Unlimited X', 'Unlimited', 'SE','Unlimited Rubican']) ]
group_config = df_Wrangler.groupby('config')
print(group_config.size().sort_values(ascending = False))
df_Wrangler['config'].value_counts().plot(kind='pie')
plt.xlabel('Configuration')
plt.ylabel('Frequency')
plt.title('Jeep Wrangler-Configuration Plot')
```

```python
## Explore the dataset: Take care of the exterior_color: ['Black', 'White', 'Brown', 'Silver', 'Red', 'Gray', 'Beige', 'Other', 'Blue']
df_Wrangler = df_Wrangler[df_Wrangler['exterior_color'].isin(['Black', 'Red', 'Silver', 'White', 'Green', 'Blue', 'Gray', 'Other', 'Yellow', 'Orange'])]
group_ext_color = df_Wrangler.groupby('exterior_color')
print(group_ext_color.size().sort_values(ascending = False))
df_Wrangler['exterior_color'].value_counts().plot(kind='pie')
plt.xlabel('Exterior Color')
plt.ylabel('Frequency')
plt.title('Jeep Wrangler-Exterior Color')
```

```python
## Explore the dataset: Take care of the interior_color: []
df_Wrangler = df_Wrangler[df_Wrangler['interior_color'].isin(['Black', 'Gray', 'Other', 'Beige'])]
group_int_color = df_Wrangler.groupby('interior_color')
print(group_int_color.size().sort_values(ascending = False))
df_Wrangler['interior_color'].value_counts().plot(kind='bar')
plt.xlabel('Interior Color')
plt.ylabel('Frequency')
plt.title('Jeep Wrangler-Interior Color')
```

```python
## Explore the dataset: Take care of the drivetrain:
df_Wrangler = df_Wrangler[df_Wrangler['drivetrain'].isin(['4x4/4-wheel drive', 'RWD'])]
group_dt = df_Wrangler.groupby('drivetrain')
print(group_dt.size().sort_values(ascending = False))
df_Wrangler['drivetrain'].value_counts().plot(kind='pie')
plt.xlabel('Drivetrain')
plt.ylabel('Frequency')
plt.title('Jeep Wrangler-Drivetrain')
```

```python
## Explore the dataset: Take care of the year:
group_yr = df_Wrangler.groupby('year')
print(group_yr.size().sort_values(ascending = False))
df_Wrangler['year'].value_counts().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Jeep Wrangler-Year')
```

```python
# scatter(x='Mileage', y='Price')
df_Wrangler.plot(kind='scatter',x='mileage', y='price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Jeep Wrangler-Price vs Mileage')
```

```python
# scatter(x='Year', y='Price')
df_Wrangler.plot(kind='scatter',x='year', y='price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Jeep Wrangler-Price vs Year')
```

```python
## Models 
# Linear Regression on Price vs Year
from sklearn.linear_model import LinearRegression
X = df_Wrangler['year'].values.reshape(-1, 1)
y = df_Wrangler['price'].values.reshape(-1, 1)
reg = LinearRegression().fit(X, y)
print(reg.coef_)
print(reg.intercept_)
price_2019 = reg.predict(np.array([2019]).reshape(-1, 1))
price_2016 = reg.predict(np.array([2016]).reshape(-1, 1))
price_2014 = reg.predict(np.array([2014]).reshape(-1, 1))
print(price_2016/price_2019)
print(price_2014/price_2019)
```

```python
# Linear Regression on Price vs Mileage
from sklearn.linear_model import LinearRegression
X = df_Wrangler['mileage'].values.reshape(-1, 1)
y = df_Wrangler['price'].values.reshape(-1, 1)
reg = LinearRegression().fit(X, y)
print(reg.coef_)
print(reg.intercept_)
price_0 = reg.predict(np.array([0]).reshape(-1, 1))
price_45k = reg.predict(np.array([45000]).reshape(-1, 1))
price_75k = reg.predict(np.array([75000]).reshape(-1, 1))
print(price_45k/price_0)
print(price_75k/price_0)
```
